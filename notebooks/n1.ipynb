```python
from src.db import get_monthly_revenue
import plotly.express as px

df = get_monthly_revenue()
print("Rows:", len(df))
fig = px.line(df, x='month', y='revenue',
              title='Monthly revenue — Jan 2022 to Dec 2024',
              labels={'revenue':'Revenue (₹)', 'month':'Month'})
fig.add_scatter(x=df['month'], y=df['revenue'].rolling(3).mean(),
                name='3-month avg', line=dict(dash='dot'))
fig.show()
```
