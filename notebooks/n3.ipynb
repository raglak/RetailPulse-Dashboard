```python
import sys, os
sys.path.append(os.path.abspath('..'))
```

```python
# Cell 2 - fast SQL version
from src.db import get_rfm_segments, get_rfm_summary
import plotly.express as px

# SQL does all RFM scoring with NTILE window functions
rfm = get_rfm_segments()
print("Rows:", len(rfm))
print(rfm[['customer_id','recency_days','frequency','monetary','rfm_segment']].head())
```

```python
# Cell 3 - segment summary table
summary = get_rfm_summary()
print(summary)
```

```python
# Cell 4 - segment distribution chart
fig1 = px.bar(summary, x='rfm_segment', y='customers',
              color='rfm_segment',
              title='Customer count by RFM segment',
              labels={'customers':'Customers','rfm_segment':'Segment'})
fig1.write_html("rfm_segments.html")
print("Saved rfm_segments.html")
```

```python
# Cell 5 - revenue by segment
fig2 = px.bar(summary, x='rfm_segment', y='total_revenue',
              color='rfm_segment',
              title='Total revenue by RFM segment',
              labels={'total_revenue':'Revenue (₹)','rfm_segment':'Segment'})
fig2.write_html("rfm_revenue.html")
print("Saved rfm_revenue.html") 
```

```python
# Cell 6 - scatter: recency vs monetary colored by segment
fig3 = px.scatter(rfm, x='recency_days', y='monetary',
                  color='rfm_segment',
                  title='RFM customer map — recency vs spend',
                  labels={'recency_days':'Days since last purchase',
                          'monetary':'Total spend (₹)'},
                  hover_data=['customer_id','city','frequency'])
fig3.write_html("rfm_scatter.html")
print("Saved rfm_scatter.html")
```
